```python
x=random.normal(size=(2,3))
print(x)
```

```python
sns.distplot(random.normal(loc=1,scale=2,size=1000),kde=True)
plt.show()
```

```python
x-random.binomial(n=8,p=0.5,size=(2,3))
print(x)
```

```python
sns.distplot(random.binomial(n=10,p=0.5,size=1000))
plt.show()
```

```python
x=random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6],size=(5))
print(x)
```

```python
sns.displot(random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6]),kde=True)
plt.show()
```
