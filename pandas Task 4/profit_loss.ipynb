```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("profit_and_loss.csv")
df
plt.figure(figsize=(10,5))
new=[]
for i in df["Profit/Loss"]:
    if i<0:
        new.append("r")
    else:
        new.append("g")

plt.bar(df["Year"],df["Profit/Loss"],zorder=2,color=new,edgecolor="k")
plt.title("Profit and Loss by Year",c="b")
plt.xlabel("Year",color="r")
plt.ylabel("Profit / Loss",color="r")
plt.grid(linestyle="--",axis="both",color="red",zorder=1)
plt.grid(which="minor",linestyle="--",axis="both",color="green",zorder=1,alpha=0.4)   
plt.minorticks_on()
plt.show()
```
