```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df=pd.read_csv('nonlinear_engine_vibration.csv')
df
```

```python
x=df['Engine RPM']
y=df['Vibration Amplitude (mm/s)']
```

```python
# y=l/1+e^-k(x-x0)
def logis_model(x, k, x0, l):
    return l/(1+np.exp(-k*(x-x0)))
```

```python
params,_=curve_fit(logis_model,x,y,p0=[0.1,600,10000])
```

```python
y_pred=logis_model(x,*params)
y_pred
```

```python
plt.scatter(x,y,label='original data')
plt.plot(x,y_pred,label='fitted curve',color='red')
plt.legend(loc='upper right',shadow=True,title_fontsize=12,title='Legend',fontsize=8)
plt.show()
```
