```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df=pd.read_csv("engine_vibration_predicted.csv")
df
```

```python
# x=df["Engine RPM"]
# y=df["Actual Vibration (mm/s)"]
X=df.drop("Actual Vibration (mm/s)",axis=1).values
y=df["Predicted Vibration (mm/s)"].values
y
```

```python
# y=ax^b
def pow_model(x,a,b):
    return a*np.power(x,b)
```

```python
params,_=curve_fit(pow_model,x,y,p0=[1,1])
```

```python
y_pred=pow_model(x,*params)
y_pred
```

```python
plt.scatter(x,y,label='original data')
plt.plot(x,y_pred,label='fitted curve',color='red')
plt.legend(loc='upper right',shadow=True,title_fontsize=12,title='Legend',fontsize=8)
plt.show()
```
