```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("students.csv")
df
```

```python
label=LabelEncoder()
for i in df.select_dtypes(include=["object"]):
    df[i]=label.fit_transform(df[i])
df
```

```python
x=df.drop("Final_Score",axis=1)
y=df["Final_Score"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
mse=mean_squared_error(y_pred,y_test)
print(f"mean_squared_error:{mse}")
rmse=np.sqrt(mse)
print(f"root_mean_squared_error:{rmse}")
score=r2_score(y_pred,y_test)
print(f"r2_score:{score}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test,y_pred,c="cyan")
plt.plot([y_test.min(),y_test.max()],[y_pred.min(),y_pred.max()],c="red")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
new_data=pd.DataFrame({
    "Study_Hours":4.37,
    "Attendance":84.91,
    "Assignments_Completed":12,
    "Department":label.fit_transform(["Arts"]),
    "Gender":label.fit_transform(["Male"])
})
predict=model.predict(new_data)
predict
plt.scatter(y_test,y_pred,c="cyan")
plt.plot([y_test.min(),y_test.max()],[y_pred.min(),y_pred.max()],c="red")
plt.plot(new_data,predict)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```
