```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv("simple_salary.csv")
df
```

```python
x=df[["Experience_Years"]]
y=df["Salary_in_Thousands"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
print(f"Mean Squared Error: {mse}")
rmse=np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
score=r2_score(y_test,y_pred)
print(f"R2 Score: {score}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color="green")
plt.plot(x_test,y_pred,color="red")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary (Thousands)")
plt.title("Experience vs. Salary")
plt.show()
```

```python
new_data=np.array([[150]])
predicted_salary=model.predict(new_data)
print(f"Predicted salary for 50 years experience: {predicted_salary[0]}")
plt.scatter(x,y,color="green")
plt.scatter(new_data,predicted_salary,color="cyan",marker="*",s=100)
plt.plot(x_test,y_pred,color="red")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary (Thousands)")
plt.title("Experience vs. Salary with New Prediction")
plt.show()
```
