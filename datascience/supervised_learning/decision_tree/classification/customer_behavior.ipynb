```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("customer_behavior.csv")
df
```

```python
label=LabelEncoder()
df["Customer_Region"]=label.fit_transform(df["Customer_Region"])
df["Age_Group"]=label.fit_transform(df["Age_Group"])
df["Prefers_Online_Shopping"]=label.fit_transform(df["Prefers_Online_Shopping"])
df["Has_Loyalty_Card"]=label.fit_transform(df["Has_Loyalty_Card"])
df["Buys_On_Discount"]=label.fit_transform(df["Buys_On_Discount"])
df["Visits_Store_Frequency"]=label.fit_transform(df["Visits_Store_Frequency"])
df["Customer_Satisfaction"]=label.fit_transform(df["Customer_Satisfaction"])
df["Preferred_Payment_Method"]=label.fit_transform(df["Preferred_Payment_Method"])
df["Recommends_To_Friends"]=label.fit_transform(df["Recommends_To_Friends"])
df["Responds_To_Offers"]=label.fit_transform(df["Responds_To_Offers"])
df["Active_Customer"]=label.fit_transform(df["Active_Customer"])

df
```

```python
x=df.drop("Active_Customer",axis=1) #features
y=df["Active_Customer"] # target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_pred,y_test)
accuracy
print(f"accuracy score:{accuracy*100:0.2f}%")
```

```python
plot_tree(model,filled=True,rounded=True,feature_names=x.columns,class_names=["Not Active_Customer","Active_Customer"])
plt.show()
```
