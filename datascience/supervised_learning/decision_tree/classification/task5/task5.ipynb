```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("weather.csv")
df
# df.isnull().sum()
```

```python
label=LabelEncoder()
df["Station.City"]=label.fit_transform(df["Station.City"])
df["Station.Code"]=label.fit_transform(df["Station.Code"])
df["Station.Location"]=label.fit_transform(df["Station.Location"])
df["Station.State"]=label.fit_transform(df["Station.State"])
df

```

```python
x=df.drop("Data.Wind.Speed",axis=1)
y=df["Data.Wind.Speed"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_pred,y_test)
accuracy
print(f"accuracy score{accuracy*100:0.2f}%")
```

```python
plot_tree(model,filled=True,rounded=True,feature_names=x.columns,class_names=["Not Data.Wind.Speed","Data.Wind.Speed"])
plt.show()
```
