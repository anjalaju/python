```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("Kmeans_800_customers.csv")
df.drop(columns="CustomerID",inplace=True)
df
```

```python
X=df
```

```python
mm=[]
kmm=range(1,11)
for k in range(1,11):
    kmodel=KMeans(n_clusters=k,random_state=42)
    kmodel.fit(X)
    mm.append(kmodel.inertia_)
mm
plt.figure(figsize=[8,6])
plt.plot(kmm,mm)
plt.show()


```

```python
model=KMeans(n_clusters=2,random_state=42)
df["clu"]=model.fit_predict(X)
```

```python
model.n_iter_
```

```python
mo=model.cluster_centers_
mo
```

```python
model.labels_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(mo[:,1],mo[:,2],c="red",marker="*",s=200)
plt.scatter(df["Annual_Income"],df["SpendingScore"],c=df["clu"])
plt.show()
```
