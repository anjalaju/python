```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift,estimate_bandwidth

```

```python
df=pd.read_csv("iot_sensor_data_1000.csv")
df
```

```python
x=df
```

```python
bandwidth=estimate_bandwidth(x,quantile=0.2)
bandwidth
```

```python
model=MeanShift(bandwidth=bandwidth)
model.fit(x)
```

```python
n=model.cluster_centers_
```

```python
m=model.labels_
```

```python
model.n_iter_
```

```python
plt.scatter(n[:,1],n[:,3],marker="*",s=200)
plt.scatter(df["Temperature_C"],df["Vibration_Level"],c=m)
plt.show()
```
