```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
# ANJAL
```

```python
df=pd.read_csv("kmean.csv")
df
```

```python
label=LabelEncoder()
for i in df.select_dtypes(include="object"):
  df[i]=label.fit_transform(df[i])
df
```

```python
x=df
```

```python
model=KMeans(n_clusters=2,random_state=42)
df["cluster"]=model.fit_predict(x)
```

```python
model.n_iter_
```

```python
# model.labels_
```

```python
model.cluster_centers_
```

```python
center_value=model.cluster_centers_
center_value
```

```python
plt.figure(figsize=(10,5))
plt.scatter(center_value[:,0],center_value[:,1],s=200,c="red",marker="*",label="centroids")
plt.scatter(df["Name"],df["m1"],c=df["cluster"],cmap="viridis",label="clusters")
plt.xlabel("Name")
plt.ylabel("m1")
plt.legend()
plt.show()
```

```python
wcss=[]
k_range=range(1,11)
for k in range(1,11):
  Kmeans=KMeans(n_clusters=k,random_state=42)
  Kmeans.fit(x)
  wcss.append(Kmeans.inertia_)
wcss
plt.figure(figsize=(10,5))
plt.plot(k_range,wcss,marker="*")
plt.xlabel("k")
plt.ylabel("wcss")
plt.title("Elbow Method")
plt.show()
```
