```python
# Prime or Not

n=int(input("Enter a number"))
if n>1:
    for i in range(2,n):
        if n%2==0:
            print("Not prime")
            break
    else:
        print(f"{n} is a prime")
else:
    print("not  a prime")
```

```python
#Odd or Even

n=int(input("Enter a number "))
if n%2==0:
    print("its Even")
else:
    print("its odd")

#Another method

for i in range(1,100):
    if i%2==0:
        print("Even")
    else:
        print("Odd")

#using function

def check_odd__even(n):
    if n%2==0:
        print("Even ")
    else:
        print("Odd")
n=int(input("Enter a number"))
check_odd__even(n)
```

```python
#multipilcation table using while loop

n=int(input("Enter a number"))
i=1
while(i<11):
    print(i,"*",n,"=",i*n)
    i=i+1
```

```python
#multipilcation table using for loop

n=int(input("Enter a number"))
for i in range(1,11):
    print(i,"*",n,"=",i*n)
    
```

```python
#factorial 

n=int(input("Enter a number"))
fact=1
while(n>0):
    fact=fact*n
    n=n-1
print(fact)
```

```python
#Reverse a number

num = 124
ans = 0
while num > 0:
    digit = num % 10      
    ans = ans *10 + digit     
    num = num // 10       

print(ans)
```
